```python
import numpy as np
import pybithumb as pb
```

```python
tickers=pb.get_tickers()[:6]
tickers
```

```python
def get_ror(k,ticker):
    df=pb.get_ohlcv(ticker)
    df['range']=(df['high']-df['low'])*k
    df['target']=df['open']+df['range'].shift(1)
    df['ma10']=df['close'].rolling(10).mean().shift(1)
    df['bull']=df['open']>df['ma10']
    fee=0.004
    df['ror']=np.where((df['high']>df['target'])&df['bull'],
                       df['close']/df['target']-fee,1)
    ror=df['ror'].cumprod()[-2]
    df['cpr']=df['ror'].cumprod()
    df['dd']=(df['cpr'].cummax()-df['cpr'])/df['cpr'].cummax()*100
    print('Times gain for','k=',k,'is',ror,', Max drawdown(%): ',df['dd'].max())
for ticker in tickers:
    result=[]
    print("Here you go ",ticker)
    for k in np.arange(0.1,1.0,0.1):
        result.append(get_ror(k,ticker))
    print(result)
```
