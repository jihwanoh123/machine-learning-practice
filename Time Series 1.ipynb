```python
#Time series data for the apple stock price
#Date-time index: Strings for now 
#Need to pass to date form(timestamp)! Done by parse_dates=['Date']
#Last method eliminates 0,1,2,... column
import pandas as pd
df=pd.read_csv('aapl.csv',parse_dates=['Date'],index_col=['Date'])
df.head(5)
```

```python
#Print all the row titles
df.index
```

```python
#Can restrict the domain into January: this is possible since we
#already converted to time-stamp
df['2017-01']
```

```python
#Average price of apple stock on January
df['2017-01'].Close.mean()
```

```python
#Can even search for a certain range
df['2017-01-07':'2017-01-01']
```

```python
#Monthly price: Choose a specific column(Close), Do resample with 
#Monthly frequence and take mean
df.Close.resample('M').mean()
```

```python
#Let's plot the chart! Monthly
%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python
#Let's plot the chart! Weekly
%matplotlib inline
df.Close.resample('W').mean().plot()
```

```python
#Quarterly frequence
df.Close.resample('Q').mean().plot()
```

```python
#Bar chart
df.Close.resample('M').mean().plot(kind='Bar')
```

```python
#Daily
df.Close.plot()
```
