```python
#Multivariable regression
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df.head()
```

```python
import math
med_bed=math.floor(df.bedrooms.median())
med_bed
```

```python
#Fill NaN with median(reaonable)
df.bedrooms=df.bedrooms.fillna(med_bed)
df
```

```python
# Creat Linear regression clss object
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df[['price']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,15]])
```

```python
#Save the regression and load it quickly
import pickle
with open('model_pickle','wb') as f:
    pickle.dump(reg,f)
with open('model_pickle','rb') as f:
    mp=pickle.load(f)
```

```python
mp.predict([[3000,3,15]])
```

```python
#Use joblib to do the same thing
from sklearn.externals import joblib
#Dump it in 'model_joblib'
joblib.dump(reg,'model_joblib')
#Load from 'model_joblib' and record it in mj
mj=joblib.load('model_joblib')
#Use mj to predict
mj.predict([[3000,3,15]])
```

```python
#Next example
from word2number import w2n
df1=pd.read_csv('hiring.csv')
df1.head()
```

```python
df1.experience=df1.experience.fillna('zero')
df1
```

```python
import math
med_score=math.floor(df1['test_score(out of 10)'].median())
df1['test_score(out of 10)']=df1['test_score(out of 10)'].fillna(0)
df1
```

```python
df1.experience=df1.experience.apply(w2n.word_to_num)
df1
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df1[['experience','test_score(out of 10)','interview_score(out of 10)']],df1[['salary($)']])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[12,10,10]])
```
