```python
#How to treat lack of uniformity in the structure of input data
#Lots of different formats: 2017-01-02, 2017/01/02, ...
#Let's uniformize
import pandas as pd
dates=['2017-01-02 2:30:00 PM','2017/01/02 14:30:00','20170102','Jan 2, 2017']
pd.to_datetime(dates)
```

```python
#How about DD/MM European type?
pd.to_datetime('5/1/2017',dayfirst='True')
```

```python
#format change
pd.to_datetime('5$1$2017',format='%d$%m$%Y')
```

```python
#In the presence of garbage strings, treat it with errors='ignore'
dates=['2017-01-02 2:30:00 PM','2017/01/02 14:30:00','20170102','Jan 2, 2017','abc']
pd.to_datetime(dates,errors='ignore')
```

```python
#In the presence of garbage strings, just coerce it to take it as 'NaT'
dates=['2017-01-02 2:30:00 PM','2017/01/02 14:30:00','20170102','Jan 2, 2017','abc']
pd.to_datetime(dates,errors='coerce')
```
