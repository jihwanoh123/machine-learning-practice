```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import time

plt.style.use(style="seaborn")
%matplotlib inline
```

```python
df=pd.read_csv('MSFT-1Y-Hourly.csv')
```

```python
X=df[['close']].copy()
X['target']=X.shift(-1)
X.dropna(inplace=True)
```

```python
def train_test_split(data,perc):
    data=data.values
    num=int(len(data)*perc)
    train=data[:num,:]
    test=data[num:,:]
    return train, test
```

```python
train,test=train_test_split(X,0.8)
```

```python
X=train[:,:-1]
y=train[:,-1]
```

```python
def xgb_predict(train,val):
    train=np.array(train)
    X,y=train[:,:-1],train[:,-1]
    model=XGBRegressor(objective='reg:squarederror',n_estimators=1000)
    model.fit(X,y)
    val=np.array(val).reshape(1,1)
    pred=model.predict(val)
    return pred[0]
```
