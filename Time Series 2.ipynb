```python
import pandas as pd
df=pd.read_csv('aapl_no_dates.csv')
df
#There is no dates!!! We want to generate dates
```

```python
rng=pd.date_range(start='06/01/2017',end='06/30/2017',freq='B')
rng
#Freq='B' is business days.
```

```python
#Use periods to specify the range
rng=pd.date_range(start='06/01/2017',periods=72,freq='B')
rng
```

```python
#Let's attach the column on the dataframe
df.set_index(rng,inplace=True)
df
```

```python
#Plot some range
%matplotlib inline
df['2017-06-01':'2017-06-10'].Close.plot()
```

```python
df.asfreq('D',method='pad') #'D' includes weekend too! not just 'B'
#If not method='pad', the weekend prices are 'NaN'
#With method='pad', we can carry the previous price to weekends.
```
