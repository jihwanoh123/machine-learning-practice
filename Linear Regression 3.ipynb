```python
#Introduce dummy variables
dummies=pd.get_dummies(df.town)
dummies.head(2)
```

```python
#Now we got dummies. Let's join two data frames
merged=pd.concat([df,dummies],axis='columns')
merged.head(3)
```

```python
#Drop 'town' and drop one of the dummies: it's causing multi-collinearity
#Why? We can deduce one from other two.
#Let's drop the last one.
final=merged.drop(['town','west windsor'],axis='columns')
final.head(3)
```

```python
#Create regression class
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
#Drop Y value and create X values
X=final.drop(['price'],axis='columns')
X.head(2)
```

```python
#Create Y 
Y=final.price
Y.head(2)
```

```python
#Get the linear regression
reg.fit(X,Y)
```

```python
# Make a prediction
reg.predict([[2800,0,1]])
```

```python
reg.predict([[3000,0,0]])
```

```python
#Score it by comparing actual value and predicted value.
reg.score(X,Y)
```
